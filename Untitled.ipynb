```python
import pandas as pd

# Replace 'your_dataset.csv' with the path to your existing CSV file
input_file = 'data/valx_indexes_50Sent.csv'

# Read the existing dataset into a DataFrame without headers
df = pd.read_csv(input_file, header=None)

# Specify the column indices you want to keep (0-indexed)
columns_to_keep = [0, 2]  # Adjust these indices as needed
new_headers = ['nctid', 'criterion']
# Create a new DataFrame with only the specified columns
new_df = df.iloc[:, columns_to_keep]
new_df.columns = new_headers
# Replace 'new_dataset.csv' with the desired output file name
output_file = 'test.csv'

# Save the new DataFrame with only the selected columns to a new CSV file
new_df.to_csv(output_file, index=False)  # No header for the new dataset

# Display the new DataFrame
print(new_df)
```
